```python
import pandas as pd
```

```python
# Load the dataset using pandas
df = pd.read_csv("day2-input.txt", header=None, skip_blank_lines=False)
```

```python
df.shape
```

```python
input_list = df[0].tolist()
```

```python
def isfloat(num):
    try:
        float(num)
        return True
    except ValueError:
        return False
```

```python
isfloat(input_list[2])
```

```python
new_list = []
my_set = set(input_list[0])
for i in range(1, len(input_list)):
    if isfloat(input_list[i]):
        if len(my_set) > 0:
            new_list.append(my_set)
        my_set = set()
    else:
        if ((len(my_set) == 0) & isfloat(input_list[i-1])):
            my_set = set(input_list[i])
        else:
            my_set = my_set.intersection(set(input_list[i]))
new_list.append(my_set)
```

```python
new_list
```

```python
total = 0
for each in new_list:
    total += len(each)
print(total)
```
